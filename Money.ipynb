### How to Filter a Pandas DataFrame on Multiple Conditions

```python
import pandas as pd

#create DataFrame

dfs = pd.DataFrame({'team': ['A', 'A', 'B', 'B', 'C'],
                   'points': [25, 12, 15, 14, 19],
                   'assists': [5, 7, 7, 9, 12],
                   'rebounds': [11, 8, 10, 6, 6]})

#view DataFrame 

dfs
```

### Example 1: Filter on Multiple Conditions Using ‘And’

```python
#return only rows where points is greater than 13 and assists is greater than 7

dfs[(dfs.points > 13) & (dfs.assists > 7)]
```

```python
#return only rows where team is 'A' and points is greater than or equal to 15

dfs[(dfs.team == 'A') & (dfs.points >= 15)]
```

### Example 2: Filter on Multiple Conditions Using ‘Or’



```python
#return only rows where points is greater than 13 or assists is greater than 7

dfs[(dfs.points > 13) | (dfs.assists > 7)]
```

```python

dfs[(dfs.team == 'A') | (dfs.points >= 15)]
```

```python
#return only rows where team is 'A' or points is greater than or equal to 15

dfs[(dfs.team == 'A') | (dfs.points >= 15)]
```

### Example 3: Filter on Multiple Conditions Using a List



```python
#define a list of values
filter_list = [12, 14, 15]

#return only rows where points is in the list of values
dfs[dfs.points.isin(filter_list)]
```

```python
#define another list of values
filter_list2 = ['A', 'C']

#return only rows where team is in the list of values
dfs[dfs.team.isin(filter_list2)]
```

### How to Exclude Columns in Pandas (With Examples)


#### exclude column1

df.loc[:, df.columns!='column1']

#### exclude column1, column2, ...

df.loc[:, ~df.columns.isin(['column1', 'column2', ...])]

```python
dfss = pd.DataFrame({'points': [25, 12, 15, 14, 19, 23, 25, 29],
                   'assists': [5, 7, 7, 9, 12, 9, 9, 4],
                   'rebounds': [11, 8, 10, 6, 6, 5, 9, 12],
                   'blocks': [2, 3, 3, 5, 3, 2, 1, 2]})

#view DataFrame
dfss
```

```python
#select all columns except 'rebounds'

dfss.loc[:, dfss.columns!='rebounds']
```

### Example 2: Exclude Multiple Columns

```python
#select all columns except 'rebounds' and 'assists'
dfss.loc[:, ~dfss.columns.isin(['rebounds', 'assists'])]
```

### How to Add a Numpy Array to a Pandas DataFrame

```python
#create pandas DataFrame
dp = pd.DataFrame({'points': [25, 12, 15, 14, 19, 23, 25, 29],
                   'assists': [5, 7, 7, 9, 12, 9, 9, 4],
                   'rebounds': [11, 8, 10, 6, 6, 5, 9, 12]})

#create NumPy array for 'blocks'
blocks = np.array([2, 3, 1, 0, 2, 7, 8, 2])

#add 'blocks' array as new column in DataFrame
dp['blocks'] = blocks.tolist()

#display the DataFrame
print(dp)
```

### Add NumPy Matrix as New Columns in DataFrame

```python
#create pandas DataFrame
ds = pd.DataFrame({'points': [25, 12, 15, 14, 19, 23]})

#create NumPy matrix
                              
mat = np.matrix([[2, 3],
                 [1, 0],
                 [2, 7],
                 [8, 2],
                 [3, 4],
                 [7, 7],
                 [7, 5],
                 [6, 3]])

#add NumPy matrix as new columns in DataFrame
df_new = pd.concat([ds, pd.DataFrame(mat)], axis=1)

#display new DataFrame
print(df_new)
```
