```python
# first table
people = {
    'first': ['Corey','Jane','John'],
    'last': ['Schafer','Doe','Doe'],
    'email':['CoreyMSchafer@gmail.com','JaneDoe@email.com','JohnDoe@email.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
# sort by last name
# can also sort numbers

df.sort_values(by = 'last', ascending = False)
```

```python
# sort multiple columns
# pass in a list of columns to sort by
# last name will be sorted first, then the first name
# depends on order

df.sort_values(by = ['last','first'], ascending = True)
```

```python
# if you want one in descending (last)
# the other one in ascending(first)
# pass a boolean in ascending list (Last = False, First = True)
# to make it permanent, use inplace = True

df.sort_values(by = ['last','first'], ascending = [False, True], inplace = True)
```

```python
df
```

```python
# set to original dataframe

df.sort_index()
```

```python
# if we want to see the sorted last name and not all

df['last'].sort_values()
```
