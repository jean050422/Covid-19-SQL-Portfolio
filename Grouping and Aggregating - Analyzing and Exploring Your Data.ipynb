```python
# use the string method
# i want all the responses from India
# when i get that results, i also want the language work with
# this just return true an false values


filt = df['Country'] == 'India'
df.loc[filt]['LanguageHaveWorkedWith'].str.contains('Python')
```

```python
# if you want to actually count
# sum can also work at booleans (true = 1, false =0)
# so there are 5233 who knows python in india

filt = df['Country'] == 'India'
df.loc[filt]['LanguageHaveWorkedWith'].str.contains('Python').sum()
```

```python
# this gives us an error
# AttributeError: 'SeriesGroupBy' object has no attribute 'str'
# use the apply method instead

country_grp['LanguageHaveWorkedWith'].str.contains('Python')
```

```python
# apply the map method instead

country_grp['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
# total number of respondents
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
# knows python

country_uses_python = country_grp['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
# combine the two dataframe above
# can combine more than one series together using concat
# axis = column will concat column
# better put the sort 

python_df = pd.concat([country_respondents,country_uses_python],axis = 'columns', sort = False )
python_df
```

```python
# rename columns

python_df.rename(columns={'Country':'NumberRespondents', 'LanguageHaveWorkedWith':'NumKnowsPython'})
```

```python
# since it looks good, put inplace = True

python_df.rename(columns={'Country':'NumberRespondents', 'LanguageHaveWorkedWith':'NumKnowsPython'},inplace = True)
```

```python
python_df
```

```python
# we have all the information that we need
# now we need to get the percentage
# (part/whole) * 100

python_df['PctKnowsPython'] = (python_df['NumKnowsPython']/python_df['NumberRespondents'])*100
```

```python
# can use in other dataframe

python_df
```

```python
# sort

python_df.sort_values(by='PctKnowsPython', ascending=False)
```

```python

python_df.sort_values(by='PctKnowsPython', ascending=False, inplace = True)
```
