```python
people = {
    "first":["Corey","Jane","John"],
    "last":["Schafer","Doe","Doe"],
    "email":["CoreyMSchafer@gmail.com","JaneDoe@email.com","JohnDoe@gmail.com"]
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df['last']=="Doe"
```

```python
# filter is a reserved keyword so do not use it as variable

filt = (df['last']== 'Doe')
```

```python
filt
```

```python
df[filt]
```

```python
# same result as above

df.loc[filt]
```

```python
# if we only need the email column of filt

df.loc[filt,'email']
```

```python
# if you want to use the & operator

filt = (df['last']== 'Doe') & (df['first']== 'John')
```

```python
df.loc[filt,'email']
```

```python
# if you want to use the OR operator which is |

filt = (df['last']== 'Schafer') | (df['first']== 'John')
```

```python
df.loc[filt,'email']
```

```python
# if you want to use the IS NOT operator which is ~ (telda)
# ~ will negate that filter and will give the opposite values

filt = (df['last']== 'Schafer') | (df['first']== 'John')
```

```python
df.loc[~filt,'email']
```
