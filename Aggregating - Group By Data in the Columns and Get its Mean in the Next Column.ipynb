```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/LinkedInLearning/data_cleaning_python_2883183/main/Ch05/05_05/cart.csv',parse_dates=['date'])
```

```python
df
```

```python
df.dtypes
```

```python
# fill the amount column missing value with value = 1

df['amount'].fillna(1, inplace=True)
df
```

```python
# take the most common name or mode and select the first one or zero
# fill the missing name with the most common name

most_common = df['name'].mode()[0]
df['name'].fillna(most_common, inplace=True)
df
```

```python
df['date'].fillna(method='ffill', inplace=True)
df
```

```python
# check two columns for groupings
df[['name','price']]
```

```python
import numpy as np
prices = df.groupby('name')['price'].transform(np.mean)
prices
```

```python
df['price'].fillna(prices, inplace=True)
df
```
